```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import tensorflow as tf
import numpy as np
```

```python
# units=units, 
dense_layer = tf.keras.layers.Dense(units=3, use_bias=False, name='my_dense_layer', activation='softmax') # activation='linear', 
```

```python
x = tf.constant([[10,10]])
dense_layer(x)
```

```python
dense_layer.get_weights()
```

```python
dense_layer.get_config()
```

```python
w = np.array([[0.5, 0.5, 0.5], [0.1, 0.1, 0.1]])
```

```python
dense_layer.set_weights([w])
```

```python
dense_layer.get_weights()
```

```python
dense_layer(x)
```

```python
dense_linear = tf.keras.layers.Dense(units=3, use_bias=False, name='my_dense_layer', activation='linear') # activation='linear',
dense_linear.build((100, 2))
```

```python
dense_linear.set_weights([w])
```

```python
dense_linear(x)
```

```python
y = tf.constant([[10,10], [10,10], [10,10]])
dense_layer(y)
```
